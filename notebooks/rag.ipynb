```python
import minsearch as ms

index = ms.Index(
    text_fields=['diet_type', 'recipe_name', 'cuisine_type', 'protein(g)', 'carbs(g)', 'fat(g)'],
    keyword_fields=[]
)

```

```python
for doc in documents:
    for key in doc:
        if isinstance(doc[key], float):
            doc[key] = str(doc[key])

index.fit(documents)
```

```python
def search(query):
    # boost = {'question': 3.0, 'section': 0.5}
    boost = {}
    number_of_results = 100

    results = index.search(
        query,
        filter_dict={},
        boost_dict=boost,
        num_results=number_of_results
    )
    return results
```

```python
entry_template = """
recipe_name: {recipe_name}
extraction_day: {extraction_day}
diet_type: {diet_type}
protein: {protein(g)}g
carbs: {carbs(g)}g
fat: {fat(g)}g
cuisine_type: {cuisine_type}
""".strip()

prompt_template = """
You are a helpful assistant for historical events. Answer the QUESTION below based on the given CONTEXT. Use the CONTEXT to provide the most accurate answer.

QUESTION: {question}

CONTEXT: {context}
"""


def build_prompt(query, results):
    context = ""
    for doc in results:
        context = context + entry_template.format(**doc) + "\n\n"

    prompt = prompt_template.format(question=query, context=context).strip()
    return prompt
```
